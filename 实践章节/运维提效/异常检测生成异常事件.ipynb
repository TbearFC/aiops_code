```python
import pandas as pd
import matplotlib.pyplot as plt
imgen_event_idt sys
import hashlib
import numpy as np

def get_md5(src):
    m2 = hashlib.md5()   
    m2.update(src.encode('utf-8'))   
    return m2.hexdigest()
def gen_event_id(x):
    if x==1:
        return get_md5(str(np.random.rand()))
    else:
        return 0
```

```python
import pandas as pd
data_path=path="../data/demo_data/metric/Twitter_volume_UPS.csv"
metric_data_df=pd.read_csv(data_path)
```

```python
data=metric_data_df["value"].values.reshape([-1,1])
data
# table = pd.concat([metric_data_df["value"]], axis=1)
# table
```

```python
import numpy as np
from sklearn.neighbors import LocalOutlierFactor
X = [[-1.1], [0.2], [10], [0.3]]
clf = LocalOutlierFactor(n_neighbors=2)
label=clf.fit_predict(data)
event=[]
for i in label:
    if i==1:
        event.append(0)
    else:
        event.append(1)
metric_data_df["event"]=event
```

```python
dt = pd.date_range(start="20210101",periods=len(metric_data_df) ,freq="5s")  # freq="D"表示频率为每一天
metric_data_df["ts"]=dt
```

```python
metric_data_df["event"]=metric_data_df["event"].apply(gen_event_id)
```

```python
metric_data_df=metric_data_df[["ts","event","value"]]
metric_data_df.to_csv("../data/demo_data/metric/metric_with_event.csv")
```
