```python
import pandas as pd
data_path=path="../data/demo_data/metric/metrics_10.csv"
metric_data_df=pd.read_csv(data_path)
metric_data_df
```

```python
from scipy.stats import pearsonr
import numpy as np
person_matrix=np.zeros([10,10])
for i in range(10):
    for j in range(10):
        person_matrix[i,j],_=pearsonr(np.array(metric_data_df[str(i)]),np.array(metric_data_df[str(j)]))
```

```python
person_matrix_df=pd.DataFrame(person_matrix)
person_matrix_df
```

```python
import seaborn as sns
%matplotlib inline
sns.set(font_scale=1.5)
sns.heatmap(data=person_matrix_df,square=True) 
```
